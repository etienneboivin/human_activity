# Human Activity Recognition with Smartphones

This dataset was built from recording 30 study participants performing activities of daily living while carrying a waist-mounted smartphone with embedded inertial sensors. The objective of this project is to classify the data into one of the six activities performed (`LAYING`, `STANDING`, `SITTING`, `WALKING`, `WALKING_UPSTAIRS`, `WALKING_DOWNSTAIRS`).

We'll use a K-nearest neighbors classifier to classify the data. Due to the large number of features (561), we'll use principal component analysis to reduce the dimensionality of the dataset. This will combat overfitting and reduce computational demands.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
def data_cleaning(df):
    temp = pd.DataFrame()
    temp["Missing Values"] = df.isnull().sum()
    temp["Unique Values"] = df.nunique()
    temp["dtype"] = df.dtypes
    return temp
```

```python
activity = joined["Activity"]
activity_counts = activity.value_counts()

plt.figure(figsize=(16, 8))
plt.bar(activity_counts.index, activity_counts)
plt.show()
```

```python
train = joined.copy()
train = joined.drop(["subject", "Activity"], axis=1)
```

PCA performs best with scaled data, so we'll use StandardScaler to standardize the dataset features to a mean of 0 and a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
train = pca.fit_transform(train)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, activity, test_size=0.2, random_state=1)
```

```python
n_folds = 10
scoring = 'accuracy'

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
model = KNeighborsClassifier()

hyperparameters = {
    "n_neighbors": range(1,50,2)
}

grid = GridSearchCV(model, param_grid=hyperparameters, cv=10)

grid.fit(train, activity)

print(grid.best_params_)
print(grid.best_score_)
```

```python
model = KNeighborsClassifier(n_neighbors=17)

model.fit(X_train, y_train)
y_predict = model.predict(X_test)

print(classification_report(y_test, y_predict))

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
plt.show()
```
